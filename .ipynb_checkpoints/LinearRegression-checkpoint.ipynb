## Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfx = pd.read_csv("./Dataset/linearX.csv")
dfy = pd.read_csv("./Dataset/linearY.csv")

x = dfx.values
y = dfy.values
Y = y[:,:]
print(x.shape)
print(y.shape)
```

```python
plt.scatter(x,y)
X = (x-x.mean())/x.std()
plt.show()
```

### Gradient Algorithm
- Start with a random theta
- Repeat until converge
- update theta according to the rule

```python
def hypothesis(x,theta):
    return theta[0]+theta[1]*x

def error(X,Y,theta):
    m = X.shape[0]
    err = 0
    
    for i in range(m):
        hx = hypothesis(X[i],theta)
        err += (hx-Y[i])**2
        
    return err

def gradient(X,Y,theta):
    grad = np.zeros((2,))
    m = X.shape[0]
    
    for i in range(m):
        hx = hypothesis(X[i],theta)
        grad[0] += (hx-Y[i])
        grad[1] += (hx-Y[i])*X[i]
        
    return grad

def gradientDescent(X,Y,learning_rate=0.001):
    theta = np.array([-2.0,0.0])
    
    itr = 0
    max_itr = 100
    error_list = []
    theta_list = []
    while(itr<=max_itr):
        grad = gradient(X,Y,theta)
        e = error(X,Y,theta)
        error_list.append(e)
        theta_list.append((theta[0],theta[1]))
        theta[0] = theta[0] - learning_rate*grad[0]
        theta[1] = theta[1] - learning_rate*grad[1]
        itr += 1
        
    return theta,error_list,theta_list
```

```python
final_theta, error_list, theta_list = gradientDescent(X,Y)
plt.plot(error_list)
plt.show()
```

#### Plot the line for testing data

```python
xtest = np.linspace(-2,6,10)
print(xtest)
```

```python
plt.scatter(X,Y,label='training data',color='blue')
plt.scatter(xtest,hypothesis(xtest,final_theta),color='red')
plt.plot(xtest,hypothesis(xtest,final_theta),color='orange',label='predicted line')
plt.legend()
plt.show()
```

## Visualizing Gradient Descent

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
T0 = np.arange(-2,3,0.01)
T1 = np.arange(-2,3,0.01)

T0,T1 = np.meshgrid(T0,T1)

J = np.zeros(T0.shape)

m = T0.shape[0]
n = T0.shape[1]

for i in range(m):
    for j in range(n):
        J[i,j] = np.sum((Y-T1[i,j]*X - T0[i,j])**2)
```

```python
theta_list = np.array(theta_list)
```

```python
## Plot the 3D Loss

fig = plt.figure()
axes = fig.gca(projection='3d')
axes.scatter(theta_list[:,0],theta_list[:,1],error_list,c='k')
axes.plot_surface(T0,T1,J,cmap='rainbow',alpha=0.5)
plt.show()
```

```python
## Contour Plot

fig = plt.figure()
axes = fig.gca(projection='3d')

axes.set_xlim([-2,2])
axes.set_ylim([-2,2])
axes.contour(T0,T1,J,cmap='rainbow')
axes.scatter(theta_list[:,0],theta_list[:,1],error_list,c='k')
plt.title("Contours in 3D")
plt.show()
```

```python
plt.contour(T0,T1,J)
plt.title("2D Contour")
plt.scatter(theta_list[:,0],theta_list[:,1],c='k',marker='*',label='Trajectory')
plt.legend()
plt.show()
```
